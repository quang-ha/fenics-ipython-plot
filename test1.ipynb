```python
from dolfin import *

mesh = UnitSquareMesh(25, 25)
Q = FunctionSpace(mesh, "CG", 1)

def boundary(x):
    return(x[1]==0 or x[0]==0)

zero = Constant(0.0)
bc = DirichletBC(Q, zero, boundary)
u = TrialFunction(Q)
v = TestFunction(Q)
f = Expression("exp(-d*pow(x[0]-x0,2) - d*pow(x[1]-y0,2)) - exp(-d*pow(x[0]-x1, 2) - d*pow(x[1]-y1,2))",
                x0=0.2, y0=0.8, x1=0.8, y1=0.2, d=100)
a = inner(grad(u), grad(v))*dx
L = f*v*dx

F = Function(Q)
solve(a == L, F, bc)
Q2 = VectorFunctionSpace(mesh, "CG", 1)
w = Function(Q2)
w = project(curl(F))
```

```python
fig = plt.figure(figsize=(20,12))
G = interpolate(f, FunctionSpace(mesh, "DG", 0))
plt.subplot(1,3,1)
p = matplot(G)
plt.colorbar(p, fraction=0.05)
plt.subplot(1,3,2)
matplot(mesh)
plt.subplot(1,3,3)
p = matplot(F)
plt.colorbar(p, fraction=0.05)
matplot(w)
plt.show()
```
